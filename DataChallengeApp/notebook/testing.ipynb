```python
import matplotlib.pyplot as plt
day = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
dataReddit = [0.5,0.2,0.3,-0.1,-0.4,0.3,0.7,0.6]
datatwitter = [0.4,0.3,0.4,-0.2,-0.5,0.1,0.5,0.7]
lineReddit = plt.plot(day, dataReddit, color='#FF4500')
lineTwitter = plt.plot(day, datatwitter, color='#1da1f2')
plt.xlabel('Day',color="green")
plt.ylabel('Sentiment',color="green")
plt.title('Sentiment Reddit/Twitter')
plt.legend(['Reddit', 'Twitter'], loc=3)
plt.axhline(y=0, color='black', linestyle='-')
plt.grid(True)
plt.show()
```

```python
import json    
from lithops import Storage
storage = Storage()
data = json.loads(storage.get_object('test-bythepego', 'twitter/2021-05-23.json'))
sentimentAvg = 0
cont = 0
for tweet in data['tweets']:
    if tweet['sentiment'] != 0:
        sentimentAvg += tweet['sentiment']
        cont += 1
sentimentAvg /= cont
print(sentimentAvg)

data = json.loads(storage.get_object('test-bythepego', 'reddit/2021-05-23.json'))
sentimentAvg = 0
cont = 0
for post in data['posts']:
    if post['sentiment'] != 0:
        sentimentAvg += post['sentiment']
        cont += 1
sentimentAvg /= cont
print(sentimentAvg)
```
